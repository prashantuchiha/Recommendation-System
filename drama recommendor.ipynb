```python
import numpy as np
import pandas as pd
import ast
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv("./data/after_preprocess.csv")
```

```python
df
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df
```

```python
def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i)
    return list
```

```python
df_copy=df
```

```python
df_copy.Genre=df_copy.Genre.apply(convert)
```

```python
type(df_copy.Genre[0])
```

```python
df_copy.Tags=df_copy.Tags.apply(convert)
```

```python
df_copy.Actors=df_copy.Actors.apply(convert)
```

```python
type(df_copy.Actors[0])
```

```python

df_copy.Synopsis=df_copy.Synopsis.apply(lambda x:x.split())
```

```python
df_copy.Synopsis
```

```python
df_copy
```

```python
df_copy.Tags=df_copy.Tags.apply(lambda x: [i.replace(" ","") for i in x])
df_copy.Genre=df_copy.Genre.apply(lambda x: [i.replace(" ","") for i in x])
df_copy.Actors=df_copy.Actors.apply(lambda x: [i.replace(" ","") for i in x])
# df_copy.Country=df_copy.Country.str.replace(" ","")
```

```python
df_copy["exp1"]=df_copy.Synopsis+df_copy.Actors+df_copy.Tags+df_copy.Genre
```

```python
df_copy.exp1[0]
```

```python
df_copy[df_copy.Title==("Vincenzo")]
```

```python
df_copy.exp1=df_copy.exp1.apply(lambda x: " ".join(x))
```

```python
df_copy.exp1
```

```python
df_copy.exp1=df_copy.exp1.apply(lambda x: x.lower())
```

```python
ps=PorterStemmer()
```

```python
def stem(obj):
    y=[]
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df_copy.exp1=df_copy.exp1.apply(stem)
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(df_copy["exp1"]).toarray()
```

```python
vector
```

```python
feature_names = cv.get_feature_names_out()

print(feature_names)
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def drama_recommend(drama):
    drama_index=df_copy[df_copy.Title==drama].index[0]
    distance=similarity[drama_index]
    drama_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
#     print(drama_list)
#     for i in drama_list:
#         print(df_copy.iloc[i[0]])
    nearest_drama = pd.DataFrame([df_copy.iloc[i[0]] for i in drama_list])
    return nearest_drama
```

```python
drama_recommend('Itaewon Class')
```
