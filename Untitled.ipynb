```python
import requests
from bs4 import BeautifulSoup

url = "https://mydramalist.com/dramalist/all"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

dramas = soup.find_all("div", class_="drama")
for drama in dramas:
    title = drama.find("a").text
    link = drama.find("a")["href"]
    print(title, link)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://mydramalist.com/shows/top"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

dramas = soup.find_all("div", class_="box")
for drama in dramas:
    drama_items = drama.find_all("li")
    for item in drama_items:
        title = item.find("a").text
        link = item.find("a")["href"]
        print(title, link)
```
